Sript by Rishi De-Kayne to produce lepmap input from VCF
FIRST- produce the sex_info_file which is produced manually the format is as follows:
column1: family number (can be all 1 if you do not have separate families)
column2: individual name 
column3: ID of father (can be blank for both parents if you only have F1s)
column4: ID of mother (can be blank for both parents if you only have F1s)
column5: sex 0 = unknown, 1 = male and 2 = female
column6: phenotype, can be 0

should be run following 01_VCF_to_LepMap_PythonINPUT.R


This python script will merge the two csvs produced side by side and reincode the genotype info from the previous R script according to the lepmap format:



```python
import pandas
import numpy
```

```python
#sls = pandas.read_csv("/home/rishi/Documents/lepmap3/PrepInput/startlines.csv")
#print(sls)
```

```python
#load sex info data 
sex_info = pandas.read_csv("sex_info_file_158.csv", header = None)
#print(sex_info)
```

```python
#load genotypes from 01 R script
newgen = pandas.read_csv("One_perLocus_trans_newgenotype_TEST158_NO_HOMOZYGOUS.csv")
print(newgen)
```

now recode genotype info 0/0 --> 1 1
                         0/1 --> 1 2
                         1/1 --> 2 2
                         ./. --> 0 0

```python
change00 = newgen.replace(to_replace="0/0", value="1 1")
```

```python
change01 = change00.replace(to_replace="0/1", value="1 2")
```

```python
change11 = change01.replace(to_replace="1/1", value="2 2")
```

```python
change = change11.replace(to_replace="./.", value="0 0")
```

```python
change.head(n=10)
change = change.drop(change.columns[[0]], axis=1)
```

```python
#now want to merge these data frames side by side to form a new input file for lepmap
lepmap_input = pandas.concat([sex_info, change], axis=1)
print(lepmap_input)
```

```python
#and to save as a tsv
lepmap_input.to_csv('lepmap_input_individualsandmarkersremoved_158_NO_HOMOZYGOUS.csv', sep = '\t', header = False, index = False )
```
